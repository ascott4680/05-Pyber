```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
df = pd.merge(city_df, ride_df, how='outer', on='city')

#reorganizing columns to be like the example
df= df[["city","date", "fare","ride_id","driver_count" , "type"]]

# Display the data table for preview
df.head()
```

```python
#grouping by city
gb_df= df.groupby(["city"])

#finding avg fare per city
avg_fare = gb_df['fare'].mean()

#finding avg rides per city
total_rides_by_city =gb_df['ride_id'].count()


#driver count by city
driver_by_city = gb_df['driver_count'].mean()

#City Types
city_type = city_df.set_index('city')['type']

#Creating dataframe to hold scatterplot data

cities = pd.DataFrame({"Average Fare":avg_fare,"Number of Rides": total_rides_by_city,
                 "Number of Drivers": driver_by_city,
                 "City Type": city_type})

cities.head()

```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = cities[cities['City Type'] == 'Urban']
suburban = cities[cities['City Type'] == 'Suburban']
rural = cities[cities['City Type'] == 'Rural']




# Build the scatter plots for each city types
plt.scatter(urban['Number of Rides'], urban['Average Fare'] , s=urban['Number of Drivers']*10, color='lightcoral', alpha= 0.90, edgecolors = "black", label="Urban")
plt.scatter(suburban['Number of Rides'], suburban['Average Fare'] , s=suburban['Number of Drivers']*10, color='Lightskyblue', alpha= 0.80, edgecolors = "black", label="Suburban")
plt.scatter(rural['Number of Rides'], rural['Average Fare'] , s=rural['Number of Drivers']*10, color='gold', alpha= 0.80, edgecolors = "black", label="Rural")


# Incorporate the other graph properties
plt.grid(alpha=0.7)
plt.xlabel('Number of Rides per City')
plt.ylabel('Average Fare per City')
plt.title('Pyber Ride Sharing Data (2016)')

# Create a legend

plt.legend(loc="upper right") #How do I made the circles the same size?

# Incorporate a text label regarding circle size
plt.text(42,40,"Note Circle Size Correlates Driver Count by City")



# Save Figure
plt.savefig("Rides by City Type.png",dpi=300)
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
#Find sum of cities

fares_df= pd.DataFrame(cities.groupby(["City Type"]).sum()['Number of Rides'])
explode = (0, 0, 0.1)
colors1=['gold','lightskyblue','lightcoral']

fares_df.plot.pie(y='Number of Rides',autopct='%1.0f%%', figsize=(7,7), explode=explode, shadow=True, colors=colors1, startangle=120, title='% of Total Fares by City Type');

# Save Figure

plt.savefig("pie_fare_by_cty_typ.png",dpi=300)
```

```python
# Show Figure
plt.show()
cities.head()
```

## Total Rides by City Type

```python
# Calculate Type Percents

#finding total cities
sum_cities= cities['City Type'].count()#120 cities


#filtering out cities into new data frames
urban_label= cities.loc[cities['City Type']=='Urban']
suburban_label= cities.loc[cities['City Type']=='Suburban']
rural_label= cities.loc[cities['City Type']=='Rural']

#counting city types
urban_count=urban_label['City Type'].count()
suburban_count = suburban_label['City Type'].count()
rural_count = rural_label['City Type'].count()

#calculating percentages
urban_percent = urban_count / sum_cities *100
rural_percent =  rural_count / sum_cities *100
suburban_percent = suburban_count / sum_cities *100

#Building dataframe to hold the above variables and values
city_pie = pd.DataFrame({ "City Type": ['Urban','Rural','Suburban'],"Percent of City Type":[urban_percent, rural_percent, suburban_percent]})

#Formatting cities as percentage
city_pie["Percent of City Type"] = city_pie["Percent of City Type"].astype(float).map("{:,.2f}%".format)

#setting index
c_df = city_pie.set_index('City Type')


#setting up my chart
yaxis = [urban_percent, rural_percent, suburban_percent]
labels =c_df.index
explode = (0.1, 0, 0)
colors=['lightcoral','gold','lightskyblue']


#showing chart
plt.pie(yaxis,labels=labels, explode=explode, startangle=90, colors=colors, autopct="%1.1f%%", shadow=True,)
plt.title('% of Total Rides by City Type')
plt.axis('equal')


# Save Figure

plt.savefig("pie_pct .png",dpi=300)

```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
drivers_df= pd.DataFrame(cities.groupby(["City Type"]).sum()['Number of Drivers'])
explode = (0, 0, 0.1)
colors2=['gold','lightskyblue','lightcoral']

drivers_df.plot.pie(y='Number of Drivers',autopct='%1.0f%%', explode=explode, shadow=True, 
                    colors=colors2, startangle=120, title='% of Total Drivers by City Type'
                   ,figsize=(7,7));

# Save Figure

plt.savefig("pie_drv_by_cty_typ.png",dpi=300)
```
