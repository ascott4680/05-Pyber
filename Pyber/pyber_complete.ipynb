```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
df = pd.merge(city_df, ride_df, how='outer', on='city')

#reorganizing columns to be like the example
df= df[["city","date", "fare","ride_id","driver_count" , "type"]]

# Display the data table for preview
df.head()
```

```python
#grouping by city
gb_df= df.groupby(["city"])

#finding avg fare per city
avg_fare = gb_df['fare'].mean()

#finding avg rides per city
total_rides_by_city =gb_df['ride_id'].count()


#driver count by city
driver_by_city = gb_df['driver_count'].mean()

#City Types
city_type = city_df.set_index('city')['type']

#Creating dataframe to hold scatterplot data

cities = pd.DataFrame({"Average Fare":avg_fare,"Number of Rides": total_rides_by_city,
                 "Number of Drivers": driver_by_city,
                 "City Type": city_type})

cities.head()

```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = cities[cities['City Type'] == 'Urban']
suburban = cities[cities['City Type'] == 'Suburban']
rural = cities[cities['City Type'] == 'Rural']




# Build the scatter plots for each city types
plt.scatter(urban['Number of Rides'], urban['Average Fare'] , s=urban['Number of Drivers']*10, color='lightcoral', alpha= 0.90, edgecolors = "black", label="Urban")
plt.scatter(suburban['Number of Rides'], suburban['Average Fare'] , s=suburban['Number of Drivers']*10, color='Lightskyblue', alpha= 0.80, edgecolors = "black", label="Suburban")
plt.scatter(rural['Number of Rides'], rural['Average Fare'] , s=rural['Number of Drivers']*10, color='gold', alpha= 0.80, edgecolors = "black", label="Rural")


# Incorporate the other graph properties
plt.grid(alpha=0.7)
plt.xlabel('Number of Rides per City')
plt.ylabel('Average Fare per City')
plt.title('Pyber Ride Sharing Data (2016)')

# Create a legend

plt.legend(loc="upper right") #How do I made the circles the same size?

# Incorporate a text label regarding circle size
plt.text(42,40,"Note Circle Size Correlates Driver Count by City")



# Save Figure
plt.savefig("Rides by City Type.png",dpi=300)
```

```python
# Show plot
plt.show()
```
